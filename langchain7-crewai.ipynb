```python
# 引入包
from langchain.chat_models import ChatOpenAI
from crewai import Agent, Task, Crew, Process
from crewai_tools import BaseTool
```

```python
from langchain.chat_models import ChatOpenAI

def get_ChatOpenAI(model_name='gpt-4o', temperature=0.0, max_tokens=None):
    return ChatOpenAI()

# 创建自定义工具
class MyCustomTool(BaseTool):
    name: str = "SimpleTextTool"
    description: str = "A tool that provides predefined text."

    def _run(self, argument: str) -> str:
        return f"Predefined text for {argument}"

# 创建自定义工具实例
simple_text_tool = MyCustomTool()
# 创建研究员代理人
researcher = Agent(
    role='高级研究员',
    goal='研究科技新趋势',
    verbose=True,
    memory=True,
    backstory=(
        "你是一位专注于发现最新科技趋势和进步的研究员。"
    ),
    tools=[simple_text_tool],
    allow_delegation=True,
    llm=get_ChatOpenAI()
)

# 创建写手代理人
writer = Agent(
    role='写手',
    goal='撰写引人入胜的科技文章',
    verbose=True,
    memory=True,
    backstory=(
        "你擅长写作，能够将复杂的信息转化为通俗易懂且引人入胜的文章。"
    ),
    tools=[simple_text_tool],
    allow_delegation=False,
    llm=get_ChatOpenAI()
)

# 定义研究任务
research_task = Task(
    description=(
        "研究AI技术的最新趋势。"
        "你的最终报告应总结关键趋势、机会和潜在风险。"
    ),
    expected_output='一份关于AI技术趋势的三段报告。',
    tools=[simple_text_tool],
    agent=researcher,
)

# 定义写作任务
write_task = Task(
    description=(
        "根据研究报告撰写文章。"
        "文章应信息丰富、引人入胜，并适合普通读者。"
    ),
    expected_output='一篇关于AI技术趋势的四段文章。',
    tools=[simple_text_tool],
    agent=writer,
    async_execution=False,
    output_file='博客文章.md'  # 输出文件
)

# 组建团队
crew = Crew(
    agents=[researcher, writer],
    tasks=[research_task, write_task],
    process=Process.sequential  # 顺序执行任务
)

# 启动任务
result = crew.kickoff(inputs={'topic': '医疗领域的AI'})
print(result)
```
