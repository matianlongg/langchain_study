```python
# 引入包
from langchain_openai import ChatOpenAI
from langchain_core.messages import HumanMessage
from langgraph.graph import END, MessageGraph
```

```python
# 定义模型
model = ChatOpenAI(
    openai_api_key='',
    openai_api_base='',
    model_name='gpt-4-0125-preview',
    temperature=0.0,
    max_tokens=None
)
```

```python
# 创建 MessageGraph 实例
graph = MessageGraph()

# 定义三个人物节点
graph.add_node("小刚", model)
graph.add_node("小明", model)
graph.add_node("小红", model)

# 连接节点形成对话循环
graph.add_edge("小刚", "小明")
graph.add_edge("小明", "小红")
graph.add_edge("小红", END)

# 设置入口点
graph.set_entry_point("小刚")

# 编译图形
runnable = graph.compile()

for output in runnable.stream(HumanMessage("嗨,大家好!")):
    for key, value in output.items():
        print(f"Output from node '{key}':")
        print("---")
        print(value)
    print("\n---\n")
```
