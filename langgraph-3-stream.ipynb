```python
from langchain_openai import ChatOpenAI
from typing import TypedDict, Optional

class State(TypedDict):
    result: Optional[str]
    counter: int

```

```python
from langgraph.graph import StateGraph, END
import time
# 创建图
graph = StateGraph(State)

# 数据分析节点
def data_analysis(state):
    # 模拟数据处理逻辑
    processed_data = "处理后的数据：" + str(state.get('counter', 0))
    state['result'] = processed_data
    state['counter'] += 1  # 累加计数器
    time.sleep(1)
    return state

graph.add_node("data_analysis", data_analysis)


# 条件节点来决定是否继续流式输出
def should_continue(state):
    # 这里假设数据流有100个数据点
    if state.get('counter', 0) >= 20:
        return END
    return "data_analysis"

graph.add_conditional_edges(
    "data_analysis", 
    should_continue, 
    {"data_analysis": "data_analysis", END: END}
)


# 设置入口点
graph.set_entry_point("data_analysis")

# 编译图形
app = graph.compile()

# 初始化状态
initial_state = State(result=None, counter=0)

# 运行图处理流式数据
state = initial_state
for s in app.stream(state):
    print(s)
```
